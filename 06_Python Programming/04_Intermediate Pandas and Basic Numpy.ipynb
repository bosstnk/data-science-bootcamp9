# Merge Dataframes

```python
left = {
    'key': [1,2,3,4],
    'name': ['toy', 'joe', 'jane', 'anna'],
    'age': [25, 28, 30, 22]
}

right = {
    'key': [1,2,3,4],
    'city': ['Bangkok', 'London', 'Seoul', 'Tokyo'],
    'zip': [1001, 2504, 2094, 9802]
}

df_left = pd.DataFrame(left)
df_right = pd.DataFrame(right)
```

```python
df_left
```

```python
df_right
```

```python
df_result = pd.merge(df_left, df_right, on='key')
df_result
```
