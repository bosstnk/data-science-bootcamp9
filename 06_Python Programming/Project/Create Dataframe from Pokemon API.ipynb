```python
## import module
from requests import get
from time import sleep
import pandas as pd

url = "https://pokeapi.co/api/v2/pokemon/"
pokemon_data = []

for i in range(10):
  index = i + 1
  url_new = url + str(index)
  response = get(url_new).json()
  sleep(2)

  pokemon_info = {
      "name": response["name"],
      "height": response["height"],
      "weight": response["weight"],
      "type_1": response['types'][0]['type']['name'] if len(response['types']) > 0 else None,
      "type_2": response['types'][1]['type']['name'] if len(response['types']) > 1 else None
  }
  pokemon_data.append(pokemon_info)

pokemon_df = pd.DataFrame(pokemon_data)
print(pokemon_df)
```
