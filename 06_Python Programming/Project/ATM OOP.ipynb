```python
from random import randint
class ATM:
  def __init__(self, account_name, bank_name, balance):
    self.account_name = account_name
    self.bank_name = bank_name
    self.balance = balance
    self.otp = None

  def __str__(self):
    return(f"This is an account of {self.account_name}, Bank: {self.bank_name}")

  def check_balance(self):
    print(f"Current balance: {self.balance}")

  def deposit(self, deposit_amount):
    if self.verify_otp():
      self.balance += deposit_amount
      print(f"Deposit amount: {deposit_amount} THB Successfully!")
      print(f"Current balance: {self.balance} THB")
      self.otp = None

  def withdraw(self, withdraw_amount):
    if self.verify_otp():
      if withdraw_amount > self.balance:
        print(f"Insufficient funds. Your balance is {self.balance} THB.")
      else:
        self.balance -= withdraw_amount
        print(f"Withdraw amount: {withdraw_amount} THB Successfully!")
        print(f"Current balance: {self.balance} THB")
      self.otp = None

  def get_OTP(self):
    self.otp = randint(1000, 9999)
    print(f"Your OTP: {self.otp} This OTP will be available in the next 2 minutes.")
    return self.otp

  def verify_otp(self):
    if self.otp is None:
      print("No OTP has been generated. Please Get OTP and Verify before making a transaction")
      return False
    entered_otp = int(input("Enter OTP: "))
    if entered_otp == self.otp:
      print("OTP verified successfully.")
      return True
    else:
      print("Invalid OTP. Please try again.")
      return False

acc1 = ATM("Boss", "A", 500)

print(acc1)
acc1.check_balance()
acc1.get_OTP()
acc1.deposit(500)
acc1.withdraw(100)
```
