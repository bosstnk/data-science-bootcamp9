```python
## import module
from requests import get
import pandas as pd
from time import sleep

## URL
url = "https://swapi.dev/api/people/"
starwars = []

## for loop
for i in range(5):
  index = i + 1 # i start with 0
  new_url = url + str(index)
  resp = get(new_url).json()
  sleep(2)

  data = {
      "name": resp["name"],
      "height": resp["height"],
      "mass": resp["mass"],
      "gender": resp["gender"],
      "homeworld": get(resp["homeworld"]).json()["name"]
  }

  starwars.append(data)

## convert to df
starwars_df = pd.DataFrame(starwars)
print(starwars_df)
## download to csv file
starwars_df.to_csv("starwars.csv")
```
