# Control Flow
- if
- for
- while

```python
# if else

score = 88

if score >= 80:
    print("A")
elif score >= 70:
    print("B")
elif score >= 60:
    print("C")
else:
    print("Failed")
```

```python
score = 85

print("passed") if score >= 80 else print("failed")
```

```python
# function
def grading(score):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    else:
        return "Failed"
grading(80)
```

```python
# for loop
artists = ["Taylor Swift",
           "Spice Girls",
           "Backstreet Boys"]

for artist in artists:
    print(artist.split(" ")[0].upper())
```

```python
# for loop
artists = ["Taylor Swift",
           "Spice Girls",
           "Backstreet Boys"]

for artist in artists:
		first_name = artist.split(" ")[0].upper()
		if first_name == "TAYLOR":
		    print("Greeeeeed! I'm going to see you in Japan.")
		else:
				print(first_name)
```

```python
# for loop
artists = ["Taylor Swift",
           "Spice Girls",
           "Backstreet Boys"]

# list comprehension
upper_artists = [a.upper() for a in artists]

print(upper_artists)
```

```python
# while loop
n = 1 # counter
while n < 5:
    print("hi!")
    n += 1 # update value for stop loop
```

# Game Function

```python
def game():
    print("This is a test game!")
    while True:
        user_input = input("Do you want to continue or stop?")
        if user_input == "stop":
            print("Game stop!")
            break
        else:
            print("Let's continue.")
```

```python
game()
```

# Object Oriented Programming

```python
class Book: # ชื่อ Class ตัวอักษาตัวแรกจะต้องเป็นตัวอักษรตัวใหญ่เท่านั้น
    def __init__(self, name, year, author):
        self.name = name
        self.year = year
        self.author = author
```

```python
book1 = Book("Think like a freak", 2010, "Dubner")
book2 = Book("Business Made Simple", 2018, "Donald Miller")
book3 = Book("Data Science for Business", 2015, "Wiley")
```

```python
book1
```

```python
book1.author
```

```python
book3.year
```

```python
class Dog:
    def __init__(self, name, age, specie):
        self.name = name
        self.age = age
        self.specie = specie

    def sitting(self):
        print("I'm sitting on the bed")

    def get_older(self, year):
        self.age += year
        print(f"I'm getting older {year} year.")
```

```python
dog1 = Dog("Andy", 3, "chihuahua")
print(dog1.name, dog1.age, dog1.specie)
```

```python
dog1.sitting()
```

```python
dog1.get_older(2)
```

```python
dog1.age
```
